import item

```python
from heapq import merge

import pandas as pd
```

fine file

```python
with open('SFSA.csv', 'r', encoding='utf-8') as f:
    raw_data = f.read()

lines = raw_data.split('\n')
lines = [line.strip() for line in lines if line.strip()]
df = pd.DataFrame(lines, columns=['data'])

# df.head()
```

empty list 


```python
comments = []
labels = []
label_ids = []
```

make it split by 3

```python
for row in df['data']:
    
    parts = str(row).split('\t')
    
    if len(parts) == 3 :
        comments.append(parts[0])
        labels.append(parts[1])
        label_ids.append(parts[2])
```

add to empty list

```python
clean_df = pd.DataFrame({
    'comment': comments,
    'label': labels,
    'label_id': label_ids
})
```

make new scv clean_file 

```python
clean_df.to_csv('clean_data.csv', index=False)

# clean_df.head()
```

fine sad comments

```python
my_list = []
for comment in clean_df["comment"][clean_df['label_id'] == "1"]:
    words = comment.split(" ")
    my_list.extend(words)
```

save sad comment to file


```python
with open('Sad_comment.txt', 'w', encoding='utf-8') as f:
    for word in my_list:
        f.write(f' {word} ')
```

make a DF for Sad_word and count

```python
WOS = pd.DataFrame({
    'word': my_list,
    'label': 1,
})
word_df_sad = WOS.groupby('word', as_index=False)['label'].sum()
word_df_sad_sorted = word_df_sad.sort_values(by='label', ascending=False).reset_index(drop=True)
word_df_sad_sorted.head(50)
```

fine happy comments


```python
my_list2 = []
for comment in clean_df["comment"][clean_df['label_id'] == "0"]:
    words = comment.split(" ")
    my_list2.extend(words)
```

save Happy comments to file

```python
with open('Happy_comment.txt', 'w', encoding='utf-8') as f:
    for word in my_list2:
        f.write(f' {word} ')
```

make a DF for Happy_word and count

```python
WDH = pd.DataFrame({
    'words': my_list2,
    'labele': 1,
})
word_df_happy = WDH.groupby('words', as_index=False)['labele'].sum()
word_df_happy_sorted = word_df_happy.sort_values(by='labele', ascending=False).reset_index(drop=True)
word_df_happy_sorted.head(50)
```

save Sad and Happy word in scv file

```python
word_df_happy_sorted.to_csv('Happy_comment.csv', index=False)
word_df_sad_sorted.to_csv('Sad_comment.csv', index=False)
```

merge to one file


```python
AC = pd.merge(word_df_happy_sorted,word_df_sad_sorted, left_on="words", right_on="word", how="outer")
all_comment_sorted = AC.sort_values(by='labele', ascending=False).reset_index(drop=True)
all_comment_sorted.to_csv('all_comment.csv', index=False)
all_comment_sorted.head(50)
```

merge to on colum


```python
AC = pd.merge(word_df_happy_sorted,word_df_sad_sorted, left_on="words", right_on="word", how="outer")
AC['total_labels'] = AC["labele"] + AC["label"]
all_comment_sorted = AC.sort_values(by='total_labels', ascending=False).reset_index(drop=True)
```

make a good name for all 
and save the file


```python
all_comment_sorted['Happy_comment_word'] = all_comment_sorted['words']
all_comment_sorted['Sad_comment_word'] = all_comment_sorted['word']
all_comment_sorted['Happy_number_use_for_word'] = all_comment_sorted['labele']
all_comment_sorted['Sad_number_use_for_word'] = all_comment_sorted['label']
all_comment_sorted['All_number_use_for_word'] = all_comment_sorted['total_labels']
```

```python
all_I_nees=['Happy_comment_word','Happy_number_use_for_word','Sad_comment_word','Sad_number_use_for_word','All_number_use_for_word']
all_comment_sorted[all_I_nees].to_csv('all_comment.csv', index=False)
```

make it more klear

```python
all_comment_sorted['%good_word']= all_comment_sorted['Happy_number_use_for_word']/all_comment_sorted['All_number_use_for_word']
all_comment_sorted['%bad_word']= all_comment_sorted['Sad_number_use_for_word']/all_comment_sorted['All_number_use_for_word']
```

join on csv

```python
my_final_list = ['Happy_comment_word','%good_word','Sad_comment_word','%bad_word','All_number_use_for_word']
all_comment_sorted[my_final_list].to_csv('%all_comment.csv', index=False)
```
